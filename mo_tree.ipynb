```python
def mylayout(node):
    if node.is_leaf():
        F = TextFace(node.org, tight_text = True)
        add_face_to_node(F, node, column=0, position='branch-right')
```

```python
ts = TreeStyle()
ts.show_leaf_name = False
```

```python
ts.layout_fn = mylayout
```
